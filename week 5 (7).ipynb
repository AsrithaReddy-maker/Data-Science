```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("tested.csv")
df.head()
```

```python
# Survival by Gender
fig, ax = plt.subplots(figsize=(6,4))

survival_gender = df.groupby("Sex")["Survived"].mean()

ax.bar(survival_gender.index, survival_gender.values)

ax.set_title("Survival Rate by Gender on the Titanic")
ax.set_xlabel("Gender")
ax.set_ylabel("Survival Rate")

plt.show()
```

```python
# Age Distribution
fig, ax = plt.subplots(figsize=(6,4))

ax.hist(df["Age"].dropna(), bins=20)

ax.set_title("Distribution of Passenger Ages")
ax.set_xlabel("Age")
ax.set_ylabel("Number of Passengers")

plt.show()
```

```python
# Survival by Class
fig, ax = plt.subplots(figsize=(6,4))

survival_class = df.groupby("Pclass")["Survived"].mean()

ax.bar(survival_class.index.astype(str), survival_class.values)

ax.set_title("Survival Rate by Passenger Class")
ax.set_xlabel("Passenger Class")
ax.set_ylabel("Survival Rate")

plt.show()
```
